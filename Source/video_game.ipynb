3.DATA CLEANING

To check missing values

```python
games.isnull().sum()
```

```python
games[games.isnull().any(axis=1)]
```

If it is True then, there are some null values found

```python
games.isnull().any()
```

To remove dupicates

```python
games.drop_duplicates(inplace=True)
```

TO FILL THE MISSING VALUES I USED FILLNA()

```python
games["Publisher"] = games["Publisher"].fillna("Unknown")
```

```python
games["Year"]=pd.to_numeric(games["Year"],errors="coerce") #It converts the 'year' column into numeric data type and the parameter tells "if a value can't converted to a number replace it with NaN"
games = games.dropna(subset=["Year"])
games["Year"]=games["Year"].astype(int)
```

5.FEATURE ENGINEERING - creating new columns from the existing data to make analysis more easier

```python
#Instead of analysing every single year,its more useful to analyze decades
#BY this we can answer for which decade sold the most games?
#and which decade released the most game
games["Decade"]=(games["Year"]//10)*10
```

To create Sales Category

```python
#Looking at exact sales values for thousands of games can be difficult. 
#Instead of that classified into categories
#pd.cut() to divide continuous num data into categories
#bins - The range(intervals) for each category
games["Sales_Category"] = pd.cut(
    games["Global_Sales"],
    bins=[0,1,5,10,60],
    labels=["Low","Medium","High","Blockbuster"]
)
```

Publisher game count - A publisher release many games but not all of them become successful

```python
publisher_games = games.groupby("Publisher")["Name"].count()
print(publisher_games)
```

7.VISUALIZATION

1.TOP PLATFORMS - Which games generated the highest global sales?

```python
platform_sales = games.groupby("Platform")["Global_Sales"].sum().sort_values()
platform_sales.plot(kind="barh", figsize=(10,6))
plt.title("Top Platforms by Global sales")
plt.xlabel("Global Sales")
plt.ylabel("Platforms")
plt.show()
```

2.TOP GENRES - Which genres are most popular?

```python
genre_sales=games.groupby("Genre")["Global_Sales"].sum()
genre_sales.plot(kind="bar")
plt.title("Genre Sales")
plt.show()
```

3.SALES TREND - How have video games sales changed over time?

```python
year_sales=games.groupby("Year")["Global_Sales"].sum()
year_sales.plot(kind="line", marker="o")
plt.title("Global Sales Trend")
plt.show()
```

4.REGIONAL SALES - Which region contributes the most sales?

```python
region_sales=games[['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
region_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Regional Sales Distribution")
plt.ylabel("")
plt.show()
```

5.TOP PUBLISHERS - Which publishers dominate the market?

```python
publisher_sales=games.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False).head(10)
publisher_sales.plot(kind="bar")
plt.title("Top Publishers")
plt.show()
```
